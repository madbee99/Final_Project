```python
data_2019 = pd.read_csv("./Resources/RawData/2019.csv")
data_2018 = pd.read_csv("./Resources/RawData/2018.csv")
data_2017 = pd.read_csv("./Resources/RawData/2017.csv")
data_2016 = pd.read_csv("./Resources/RawData/2016.csv")
data_2015 = pd.read_csv("./Resources/RawData/2015.csv")
data_2014 = pd.read_csv("./Resources/RawData/2014.csv")
data_2013 = pd.read_csv("./Resources/RawData/2013.csv")
data_2012 = pd.read_csv("./Resources/RawData/2012.csv")
data_2011 = pd.read_csv("./Resources/RawData/2011.csv")
data_2010 = pd.read_csv("./Resources/RawData/2010.csv")
data_2009 = pd.read_csv("./Resources/RawData/2009.csv")
value_labels = pd.read_csv("./Resources/Value_Labels.csv")
```

```python
new_col_names = list(['UnitID', 'SchoolName', 'StateCode',
       'Sector', 'OpenAdmissionPolicy',
       'GPA', 'SchoolRank',
       'SecondarySchoolRecord', 'Recs',
       'ApplicantsTotal', 'ApplicantsMen',
       'ApplicantsWomen', 'AdmissionsTotal',
       'AdmissionsMen', 'AdmissionsWomen',
       'EnrolledTotal', 'EnrolledMen',
       'EnrolledWomen',
       'SAT_RW_25',
       'SAT_RW_75',
       'SAT_Math_25',
       'SAT_Math_75',
       'ACT_Composite_25',
       'ACT_Composite_75',
       'ACT_English_25',
       'ACT_English_75',
       'ACT_Math_25',
       'ACT_Math_75', 'SchoolYear'])
new_col_names
```

```python
clean_cols_2019 = data_2019.set_axis(new_col_names, axis = 1)
clean_cols_2018 = data_2018.set_axis(new_col_names, axis = 1)
clean_cols_2017 = data_2017.set_axis(new_col_names, axis = 1)
clean_cols_2016 = data_2016.set_axis(new_col_names, axis = 1)
clean_cols_2015 = data_2015.set_axis(new_col_names, axis = 1)
clean_cols_2014 = data_2014.set_axis(new_col_names, axis = 1)
clean_cols_2013 = data_2013.set_axis(new_col_names, axis = 1)
clean_cols_2012 = data_2012.set_axis(new_col_names, axis = 1)
clean_cols_2011 = data_2011.set_axis(new_col_names, axis = 1)
clean_cols_2010 = data_2010.set_axis(new_col_names, axis = 1)
clean_cols_2009 = data_2009.set_axis(new_col_names, axis = 1)
```

```python
all_data = pd.concat([clean_cols_2019, clean_cols_2018,clean_cols_2017,clean_cols_2016,clean_cols_2015,clean_cols_2014,clean_cols_2013,clean_cols_2012,clean_cols_2011,clean_cols_2010,clean_cols_2009])
all_data = all_data.sort_values(by = ['SchoolName', 'SchoolYear'])
all_data
```

```python
all_data.dtypes
```

```python
all_data["Sector"]=all_data["Sector"].replace({1: "Public", 2: "Private"})
all_data
```

```python
all_data_clean2 = all_data.dropna()
all_data_clean2
```

```python
all_data_clean2["OpenAdmissionPolicy"]=all_data_clean2["OpenAdmissionPolicy"].replace({1: "Yes", 2: "No", -2: "N/A"})
all_data_clean2
```

```python
all_data_clean2[["GPA","SchoolRank", "SecondarySchoolRecord", "Recs"]]=all_data_clean2[["GPA","SchoolRank", "SecondarySchoolRecord", "Recs"]].replace({1: "Required", 5: "Considered but NOT required", 2: "Recommended", 3: "Not Required nor Recommended"})
all_data_clean2
```

```python
counts = all_data_clean2['UnitID'].value_counts()
counts
```

```python
# Remove colleges for which we don't have all years of data
to_remove = counts[counts < 11].index
to_remove
```

```python
all_data_clean3 = all_data_clean2[~all_data_clean2.UnitID.isin(to_remove)]
all_data_clean3
```

```python
all_data_clean3 = all_data_clean3.reset_index(drop = True)
all_data_clean3 
```

```python
all_data_clean3.dtypes
```

```python
all_data_clean3.to_csv('./CleanData/cleaned_data_numerical_state_code.csv', index=False)  
```
